```python
# 學生總分為300分
# 有些學生可以加分5%

try:
    scores = int(input("請輸入學生分數(最高300分):"))
    print(scores)
except:
    print("出錯了")

scores = int(input("請輸入學生分數(最高300分):"))
is_add = input("學生是否符合加分條件?(y,x)")

if is_add == "y": #單項選擇
    scores = scores * 1.05

# 四捨五入(5進位)
scores = int(scores + 0.5)

# 超過300分，上限取300分
if scores > 300:
    scores = 300

print("學生分數:",scores)
```
