```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors = ['yellow', 'green', 'red', 'blue']

figure = plt.figure()
axes = figure.add_subplot()
axes.pie(values,
         labels=labels,
         colors=colors,
         explode=[0.3, 0, 0, 0], #把一塊往外拉開
         shadow=True, #增加陰影
         autopct="%2.1f%%", #顯示百分比
         startangle=180) #旋轉角度
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['milk tea', 'black tea', 'green tea', 'coffee']
values = [50, 25, 35, 40]
colors = ['yellow', 'red', 'green', 'brown']

figure = plt.figure()
axes = figure.add_subplot()

axes.pie (values, 
        labels=labels, 
        colors=colors, 
        explode=[0.2, 0, 0, 0], 
        shadow=True, 
        startangle=90, 
        autopct='%1.1f%%')

plt.show()
```
