```python
import pandas as pd

dataFrame = pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame.dtypes #查看 DataFrame 每一欄的資料型別的屬性
len(dataFrame)
dataFrame.info() #查看 DataFrame 結構摘要
```

```python
#改變欄位的名稱
dataFrame.columns = ["統計年", "區域別", "年底人口數", "土地面積", "人口密度"]

#刪除第0列
df = dataFrame.drop(index = 0)

#清除掉有NaN的欄位
df1 = df.dropna()

#刪除「統計年」欄位，1代表欄，0代表列
df2 = df1.drop('統計年', axis=1)

#將「年底人口數」強制轉為數字，像 '…' 這樣的文字會被強制變成 NaN
df2["年底人口數"] = pd.to_numeric(df2["年底人口數"], errors="coerce")
df2

#這時候再用 dropna()，就能把原本是 '…' 的東沙、南沙群島（以及其他原本就是 NaN 的列）一次清乾淨！
df2 = df2.dropna(subset=["年底人口數"])
df2.info()

#最後轉換為整數
df2["年底人口數"] = df2["年底人口數"].astype("int64")
df2["土地面積"] = df2["土地面積"].astype("float")
df2["人口密度"] = df2["人口密度"].astype("int64")
df2
```
