```python
#引數值的呼叫 open(”student.txt”) 必須依順序

try:
	with open("student.txt", "r", -1, encoding="utf-8") as file: 
			#r代表讀取模式
			#buffering=-1 通常表示使用系統預設的緩衝設定
            #encoding="utf-8" 代表用 UTF-8 編碼讀檔
			
			data = file.read()
			print(data)
except Exception as e:
	print("發生錯誤:", e)
```

```python
#引數名稱的呼叫 open(file=”student.txt”) 可以不依順序，只要依照名稱

try:
	with open(file="student.txt", encoding="utf-8") as file: 
			data = file.read()
			print(data)
except Exception as e:
	print("發生錯誤:", e)
```

```python
# 混合名稱的呼叫 open(file="student.txt") 可以不依順序,只要依照名稱

try:
	with open("student.txt",encoding="utf-8",mode="r") as file:
			data = file.read()
			print(data)
except Exception as e:
	print("發生錯誤:", e)
```
